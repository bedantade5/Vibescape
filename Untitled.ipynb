### 2. Data Cleaning
### --------------------------------------------------

```python
# Checks if there is any missing values
print("Missing Values Count:")
print(df.isnull().sum())

# Handle missing values if any
if df.isnull().sum().sum() > 0:
    print("\nHandling missing values...")
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numerical_cols:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].mean())
    
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].mode()[0])
    
    print("Missing values handled.")
else:
    print("No missing values found.")

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

if duplicates > 0:
    print("Removing duplicate rows...")
    df = df.drop_duplicates()
    print(f"Dataset now has {df.shape[0]} rows after removing duplicates.")
```

### 5. Simple Recommendation Prototype
### --------------------------------------------------


Simple prototype for recommending songs based on mood
    
Parameters:
mood (str): One of 'happy', 'sad', 'energetic', 'relaxed'
n_recommendations (int): Number of recommendations to return
    
Returns:
DataFrame: Top recommended songs for the given mood
    

```python
def recommend_songs_by_mood(mood, n_recommendations=5):
    
    mood_songs = df[df['mood'] == mood]
    
    if len(mood_songs) == 0:
        return "No songs found for this mood"
    
    recommendations = mood_songs.sort_values('popularity', ascending=False).head(n_recommendations)
    
    return recommendations[['track_name', 'artists', 'album_name', 'popularity', 'track_genre']]

for mood in ['happy', 'sad', 'energetic', 'relaxed']:
    print(f"\nTop recommended songs for '{mood}' mood:")
    print(recommend_songs_by_mood(mood))
```

### 7. Content-Based Recommendation Approach
### --------------------------------------------------

Content-based recommendation using cosine similarity of audio features
    
Parameters:
song_id (str): Track ID of the song to base recommendations on
n_recommendations (int): Number of recommendations to return
    
Returns:
DataFrame: Similar songs based on audio features

```python
def get_content_based_recommendations(song_id, n_recommendations=5):
    
    if song_id not in df['track_id'].values:
        return "Song not found in dataset"
    
    song_idx = df[df['track_id'] == song_id].index[0]
    
    features_matrix = df[audio_features].values
    similarity_scores = cosine_similarity([features_matrix[song_idx]], features_matrix)
    
    similar_indices = similarity_scores[0].argsort()[::-1][1:n_recommendations+1]
    
    recommendations = df.iloc[similar_indices][['track_name', 'artists', 'album_name', 'popularity', 'track_genre', 'mood']]
    return recommendations

if len(df) > 0:
    random_song_id = df['track_id'].sample(1).values[0]
    song_info = df[df['track_id'] == random_song_id][['track_name', 'artists', 'mood']].values[0]
    
    print(f"Test song: {song_info[0]} by {song_info[1]} (Mood: {song_info[2]})")
    print("\nSimilar songs based on audio features:")
    print(get_content_based_recommendations(random_song_id))
```

### 8. Proposed Mood-Based Recommendation System
### --------------------------------------------------

Enhanced mood-based recommendation system
    
Parameters:
selected_mood (str): User's selected mood ('happy', 'sad', 'energetic', 'relaxed')
n_recommendations (int): Number of recommendations to return
    
Returns:
DataFrame: Recommended songs for the selected mood

```python
def get_mood_based_recommendations(selected_mood, n_recommendations=5):
    mood_songs = df[df['mood'] == selected_mood]
    
    if len(mood_songs) == 0:
        return "No songs found for this mood"
    
    # Calculate a composite score based on popularity and audio feature relevance
    # For each mood, we want to prioritize different audio features:
    if selected_mood == 'happy':
        # Happy: High valence, moderate-high energy, high danceability
        mood_songs['mood_score'] = (
            mood_songs['valence'] * 0.5 + 
            mood_songs['energy'] * 0.3 + 
            mood_songs['danceability'] * 0.2
        )
    elif selected_mood == 'sad':
        # Sad: Low valence, low energy, high acousticness
        mood_songs['mood_score'] = (
            (1 - mood_songs['valence']) * 0.5 + 
            (1 - mood_songs['energy']) * 0.3 + 
            mood_songs['acousticness'] * 0.2
        )
    elif selected_mood == 'energetic':
        # Energetic: High energy, high tempo, high danceability
        mood_songs['mood_score'] = (
            mood_songs['energy'] * 0.5 + 
            mood_songs['tempo'] / 200 * 0.3 +
            mood_songs['danceability'] * 0.2
        )
    elif selected_mood == 'relaxed':
        # Relaxed: Low energy, high acousticness, moderate valence
        mood_songs['mood_score'] = (
            (1 - mood_songs['energy']) * 0.4 + 
            mood_songs['acousticness'] * 0.4 + 
            (1 - abs(mood_songs['valence'] - 0.5)) * 0.2
        )
    
    # Combine mood score with popularity for final ranking
    mood_songs['final_score'] = mood_songs['mood_score'] * 0.7 + mood_songs['popularity'] / 100 * 0.3
    
    # Sort by final score and return top recommendations
    recommendations = mood_songs.sort_values('final_score', ascending=False).head(n_recommendations)
    
    return recommendations[['track_name', 'artists', 'album_name', 'popularity', 'track_genre']]

for mood in ['happy', 'sad', 'energetic', 'relaxed']:
    print(f"\nTop recommended songs for '{mood}' mood (enhanced method):")
    print(get_mood_based_recommendations(mood))
```
